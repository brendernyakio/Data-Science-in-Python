```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
import warnings
warnings.simplefilter(action='ignore')
```

```python
class_boundaries = [299.5, 309.5, 319.5, 329.5, 339.5,359.5]
frequencies = [4, 6, 10, 1, 5,4]

midpoints = [304.5, 314.5, 324.5, 334.5,344.5, 354.5]
polygon_frequencies = [4,6,10,1,5,4]

midpoints_extended = [294.5] + midpoints +[364.5]
frequencies_extended = [0] + polygon_frequencies + [0]

plt.figure()

for i in range(len(frequencies)):
    plt.bar(class_boundaries[i], frequencies[i], 
            width=10, align='edge')

plt.plot(midpoints_extended, frequencies_extended, marker='o') 

plt.xlabel("Class Intervals")
plt.ylabel("Frequency")
plt.title("Histogram and Frequency Polygon")

plt.show()
```

# Histogram:Income Distribution

```python
sns.histplot(gss, x='income', bins=30, kde=True, color="red")
plt.title("Histogram Showing the Distribution of Income")
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.show()
```

# Boxplot:Income by Degree

```python
sns.boxplot(data=gss, x='degree',y='income', palette='Set1') 
plt.title(" Income by Degree")
plt.xlabel("Degree")
plt.ylabel("Income")
plt.show()
```

# Customizing Scatter Plot With Color and Size

```python
sns.scatterplot(data=gss, x='age',y='income',hue='sex', size='income', alpha=0.6, palette=['red','blue']) 
plt.title(" Income vs Age by Gender")
plt.xlabel("Age")
plt.ylabel("Income")
plt.legend(title='Gender')
plt.show()
```

```python
sns.scatterplot(data=gss, x='age',y='income',hue='sex', size='income', alpha=0.9, palette=['red','blue']) 
plt.title(" Income vs Age by Gender")
plt.xlabel("Age")
plt.ylabel("Income")
plt.legend(title='Gender')
plt.show()
```

# ADVANCED CHART
## Scatter Plot with Regression Line

```python
plt.figure(figsize=(8,6))
sns.regplot(x='age', y='income', data=gss, scatter_kws={'alpha':0.5}, line_kws={"color":"red"})
plt.title("Income vs Age (Regression Line)")
plt.xlabel("Age")
plt.ylabel("Income")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.regplot(x='age', y='income', data=gss, scatter_kws={'alpha':0.9}, line_kws={"color":"purple"})
plt.title("Income vs Age (Regression Line)")
plt.xlabel("Age")
plt.ylabel("Income")
plt.show()
```

# Violin Plot

```python
plt.figure(figsize=(4,3))
sns.violinplot(x='sex', y='income', data=gss, palette='Set2')
plt.title("Income Distribution by Gender (Violin Plot)")
plt.xlabel("Gender")
plt.ylabel("Income")
plt.grid(True)
plt.show()
```

# Heat Map(Correlation Matrix)

```python
plt.figure(figsize=(10,6))
corr_matrix = gss.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```
