## **Two Sample T-Test**
### A two-sample t-test compares the means of two independent groups to determine if they differ significantly, assuming normal distribution and similar variances.

### a).Independent Two-Sample t-Test(Age by Gender)

```python
#H0:There is no significant difference between average age across gender.
#H1:There is significant difference between average age across gender.

group_A = gss[gss['sex'] == 'MALE']['age'].dropna()
group_B = gss[gss['sex'] == 'FEMALE']['age'].dropna()

two_sample_test = pg.ttest(group_A,group_B, paired=False)
two_sample_test
```

### b).Independent Two-Sample t-Test(Gender by Income)

```python
#H0:There is no significant difference between average income across gender.
#H1:There is significant difference between average income across gender.

group_A = gss[gss['sex'] == 'MALE']['income'].dropna()
group_B = gss[gss['sex'] == 'FEMALE']['income'].dropna()

two_sample_test = pg.ttest(group_A,group_B, paired=False)
two_sample_test
```

```python
#H0:There is no significant difference between average hrswrk across gender.
#H1:There is significant difference between average hrswrk across gender.

group_A = gss[gss['sex'] == 'MALE']['hrswrk'].dropna()
group_B = gss[gss['sex'] == 'FEMALE']['hrswrk'].dropna()

two_sample_test = pg.ttest(group_A,group_B, paired=False)
two_sample_test
```

####  Since P-Value < 0.05, we reject H0 and conclude that there is significant difference between average hrswrk across gender.


## **Paired t-Test**
### A paired sample t-test compares the means of two related groups(eg..,before-and-after measurements on the same subjects) to determine if the average difference between paired observations is statistically significant.

```python
#H0:There is no significant difference on average test grades.
#H1:There is significant difference on average test grades.

before = [72,75,78,71,69,74,77,73,76,70]
after = [75,78,80,74,72,77,79,75,78,73]


# perform paired t-test
t_statistic, p_value = stats.ttest_rel(before, after)

# Output results
print("t-statistic:", t_statistic)
print("p-value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: there is no significant difference on average test grades.")
```

## Alternative

```python
pg.ttest(before, after, paired = True)
```

## Chi-Square Test for Independence

```python
#H0:The two variables are independent (no relationship exists)
#H1:The two variables are dependent (there is a significant relationship)

contingency_table = pd.crosstab(gss["sex"],gss["marital"])

chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

print("Chi-Square Test Result(Gender vs Marital Status):")
print("Chi2:", chi2, "p-value:",p,"Degrees of Freedom:", dof)
print("p-value:", p)
print("Degree of Freedom:", dof)

alpha = 0.05
if p< alpha:
    print("Reject the null hypothesis: there is a significant association between the variables.")
else:
    print("Fail to reject the null hypothesis: no significant association between the variables.")
```
