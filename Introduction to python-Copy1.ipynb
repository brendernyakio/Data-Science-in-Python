## **Magnitude of a vector**

```python
magnitude_a = np.linalg.norm(vector_a)
magnitude_a
print("magnitude of a:", magnitude_a)
rounded_value = round(magnitude_a, 3)
print("magnitude of a:", rounded_value)
rounded_value = round(magnitude_a, 2)
print("magnitude of a:", rounded_value)

rounded_value = round(magnitude_a, 1)
print("magnitude of a:", rounded_value)
```

# **Matrices**

### **Defn- a Matrix is a 2D array of numbers arranged in rows and column**

#### **Quick Practice Exercise**

```python
import numpy as np

# Create two 2x2 matrices
A = np.array([[1,2],
              [3,4]])
B = np.array([[2,0],
              [1,2]])
```

```python
# Perform operations
print("A + B =\n", A + B)
```

```python
print("\nA - B =\n", A - B)
```

```python
print("\nA X B (element-wise) =\n", A * B)
```

```python
print("\nA@ B (matrix multiplication) =\n", A @ B)
```

```python
print("\nTranspose of A =\n", A.T)
```
