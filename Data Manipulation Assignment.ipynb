## EXERCISE D

```python
import pandas as pd

# Load your data (adjust file path if needed)
df = pd.read_csv('german_credit__data.csv')


# --- 1. Average value per category (e.g., 'Job') ---
avg_per_category = df.groupby('Job')['Credit amount'].mean().reset_index()
avg_per_category = avg_per_category.rename(columns={'Credit amount': 'Average Credit amount'})

# --- 2. Count per category ---
count_per_category = df.groupby('Job').size().reset_index(name='Count')

# --- 3. Merge both results for easier sorting ---
summary = pd.merge(avg_per_category, count_per_category, on='Job')

# --- 4. Sort by average value descending ---
summary_sorted = summary.sort_values(by='Average Credit amount', ascending=False)

# Show final result
print("\nAverage Credit amount per Job (sorted descending):")
summary_sorted
```

## EXERCISE E

```python
import pandas as pd

# 1. SETUP: Create the pipeline (wrapped in parentheses for multi-line chaining)
clean_analysis = (
    df
    # VERB 1: 'drop' - Remove unnecessary columns to keep the result "clean"
    .drop(columns=['S/no'])
    
    # VERB 2: 'groupby' - Segment the data by the borrower's 'Purpose'
    .groupby('Purpose')
    
    # VERB 3: 'agg' - Summarize with specific names for the output columns
    .agg(
        average_loan=('Credit amount', 'mean'),
        average_months=('Duration', 'mean'),
        applicant_count=('Age', 'count')
    )
    
    # VERB 4: 'sort_values' - Rank the results by the largest average loan amount
    .sort_values(by='average_loan', ascending=False)
)

# Display the final, polished table
clean_analysis
```
