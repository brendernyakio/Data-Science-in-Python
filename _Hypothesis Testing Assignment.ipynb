## Exercise D: Chi-Square Test
### Select two categorical variables (e.g., occupation vs payment_type )
### Build a contingency table
### Perform a chi-square test for independence
### Interpret the result

```python
import pandas as pd
from scipy import stats


contingency_table = pd.crosstab(gss["occupation"], gss["payment_type"])

# perform chi-square test
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

print("Chi-Square Test Result (Occupation vs Payment_type):")
print("Chi2:", chi2, "p-value:", p, "Degrees of Freedom:", dof)
print("p-value:", p)
print("Degrees of Freedom:", dof)

# Interpret the result
alpha = 0.05
if p < alpha:
    print("Reject the null hypothesis: there is a significant association between the variables.")
else:
    print("Fail to reject the null hypothesis: no significant association between the variables.")
```
